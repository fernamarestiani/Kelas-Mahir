Perceptron adalah salah satu metode Jaringan Syaraf Tiruan (JST) sederhana yang menggunakan algoritma training untuk melakukan klasifikasi secara linier. Perceptron terdiri dari Dendrit, Nukleus, Badan sel, axon, dan axon terminal.Perceptron digunakan untuk melakukan klasifikasi sederhana dan membagi data untuk menentukan data mana yang masuk dalam klasifikasi dan data mana yang missclasifikasi (diluar klasifikasi). Perceptron dapat kita gunakan untuk memisahkan data yang dapat kita bagi menjadi 2 kelas, misalnya kelas C1 dan kelas C2.

Perceptron dalam Jaringan Syaraf Tiruan memiliki bobot yang bisa diatur dalam suatu ambang batas (threshold). Melalui proses pembelajaran (learning), Algotirma Perceptron akan mengatur parameter-parameter bebasnya. 

Intinya pada perceptron terjadi proses dimana sel menerima input, memprosesnya kemudian menghasilkan output.

## Pengenalan Tensorflow


**Model Neural Networks Dasar**

Tensorflow adalah sebuah library pada python untuk Deep Learning. pada bagian ini akan fokus pada syntaks bagaimana melakukan pemodelan Neural Network menggunakan library TensorFlow python. pada bagian ini kita akan mulai pada kasus paling sederhana yaitu adalah Tabel Logika "And" sebagai berikut

```
x1   x2   y
1    1    1
1    0    0
0    1    0
0    0    0
```
perhatikan bahwa x1, dan x2 adalah input, dan y adalah output yang diinginkan. Maka **ukuran input adalah dua kolom dan 4 baris sedangkan input adalah sebuah nilai antara 0 dan 1.**


````python
# import library
import tensorflow as tf
from tensorflow import keras
import numpy as np
````

````python
# Menampilkan versi tensorflow
tf.__version__
````

Tiga tahap penting:

1.   Menyusun Arsitektur Model
2.   Mengkompilasi Model
3.   Training Model
4.   Evaluasi Model

````python
# Menyusun model
model = keras.Sequential([                              # Membuat model sequential
    tf.keras.layers.Dense(units=1,
                          input_shape=[2],
                          activation='relu'),           # Menambah layer pada model NN             
])

# Compile model
model.compile(loss='mean_squared_error',                # Menentukan fungsi loss
              optimizer='sgd')                          # Menentukan fungsi optimasi

model.summary()                                         # Menampilkan summary model
````

Berdasarkan model summary diatas maka kita sedang membuat model NN dengan satu titik layer. Kemudian hasil parameternya adalah tiga. itu berarti terdapat toga parameter, apa aja? yaitu w1, w2 karena inputnya adalah x1 dan x2, serta b yaitu bias. berikut adalah visualisasi model NN nya

WhatsApp Image 2021-12-08 at 16.10.29.jpeg

````python
# Mendefinisikan data input (X) dan output (Y)
x = np.array([[1,1], [1,0], [0,1], [0,0]], dtype="int")
y = np.array([[1], [0], [0], [0]], dtype="int")
````

````python
# Menampilkan data x
x
````

````python
# Menampilkan data y
y
````

````python
# Melatih model (fitting/training)
history = model.fit(x, y, epochs=2000, verbose=2)
````

````python
# Prediksi model
model.predict(x)
````

````python
# Membulatkan hasil pemodelan
model.predict(x).round()
````

````python
# Menampilkan nilai bobot
model.get_weights()
````
